```python
import os
import numpy as np
import collections
import itertools
```

```python
with open('/Users/januaryshen/Dropbox/S19 - CSE 547/hw1/q2/data/browsing.txt') as myfile:
    file_all = myfile.readlines()
```

```python
item_dict = {}
location = {}
for i, line in enumerate(file_all, start = 0):
    for j in line.split(sep = ' '):
        item_dict[j] = item_dict.get(j, 0) + 1
        
        if j in location.keys():
            location[j].append(i)
        else:
            location[j] = [i]
```

```python
frequent_dict = {k:v for (k,v) in item_dict.items() if v >= 100}
#drop \n in frequent_dict, there are 31101 \n's
frequent_dict.pop('\n', None)
```

```python
conf_dict = {}
keys = list(frequent_dict)
location_tuple = {}

for i in range(0, len(keys)):
    for j in range(i+1, len(keys)):
        intersection = len(set(location[keys[i]]) & set(location[keys[j]]))
        conf_dict[(keys[i], keys[j])] = intersection/frequent_dict[keys[j]]
        location_tuple[(keys[i], keys[j])] = intersection
```

```python
res = sorted(conf_dict.items(), key=lambda x: x[1], reverse = True)
res[0:5]
```

```python
# triple
frequent_dict_tuple = {k:v for (k,v) in location_tuple.items() if v >= 100}
keys_tuple = list(frequent_dict_tuple)
the_extra = list()

for i in range(0, len(keys_tuple)):
    if keys_tuple[i][0] not in the_extra:
        the_extra.append(keys_tuple[i][0])
    if keys_tuple[i][1] not in the_extra:
        the_extra.append(keys_tuple[i][1]) 
```

```python
conf_dict_triple = {}
for i in range(0, len(keys_tuple)): #len(keys_tuple)
    for j in range(0, len(the_extra)): #len(the_extra)
        if the_extra[j] not in keys_tuple[i]:
            conf_dict_triple[(keys_tuple[i][0],keys_tuple[i][1],the_extra[j])] = len(set(location[keys_tuple[i][0]]) & set(location[keys_tuple[i][1]]) & set(location[the_extra[j]]))/len(set(location[keys_tuple[i][0]]) & set(location[keys_tuple[i][1]]))
            
```

```python
res_triple = sorted(conf_dict_triple.items(), key=lambda x: x[1], reverse = True)
res_triple
```

```python
temp = res_triple[0:15]
temp_list = []

for i in range(0,15):
    temp_list.append([sorted(temp[i][0][0:2]),temp[i][0][2]])

temp_list.sort()
```
